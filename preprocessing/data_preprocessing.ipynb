```python
import pandas as pd
df = pd.read_csv('data_sst.csv')
df.head()
```

```python
df = df.drop(['Restaurant', 'Reviewer', 'Metadata', 'Time', 'Pictures','7514'], axis=1)
df.head()
```

```python
df = df.rename(columns={'Review': 'sentence', 'Rating': 'sentiment'})
df.head()
```

```python
# seeing which values are in the sentiment column
df['sentiment'].value_counts()
```

```python
df = df[df['sentiment'] != '1.5']
df = df[df['sentiment'] != '2.5']
df = df[df['sentiment'] != '3.5']
df = df[df['sentiment'] != '4.5']
df = df[df['sentiment'] != 'Like']
```

```python
df['sentiment'].value_counts()
```

```python
df.info()
```

```python
# removing rows which does not have sentence
df = df.dropna(subset=['sentence'])
df.info()
```

```python
#making extra column with id
df['id'] = range(1, len(df) + 1)
df.head()
```

```python
#making id the first column
cols = df.columns.tolist()
cols = cols[-1:] + cols[:-1]
df = df[cols]
df.head()
```

```python
#saving df to csv
df.to_csv('data_sst.csv', index=False)
```
